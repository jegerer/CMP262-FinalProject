**Question 1: Based on average ratings for breweries, which breweries perform the best and the worst overall?**

Before conducting this analysis, we should consider trimming down the breweries that have only a small handful of different products.  To this end, an arbitrary threshold of five is chosen.

```python
breweries_more_than_5 = df.groupby('brewery').filter(lambda x : x['brewery'].size >= 5)
```

```python
breweries_more_than_5.groupby('brewery')['name'].count().sort_values(ascending = False)
```

By choosing this threshold, the number of remaining breweries has been reduced from 1,498 to 309.

```python
brew_more_5_sorted = breweries_more_than_5.groupby('brewery').aggregate({'brewery': 'count', 'avg_rating': 'mean'}).sort_values(by = 'avg_rating', ascending = False)
brew_more_5_sorted
```

```python
breweries_more_than_5
```

```python
brew_more_5_sorted.rename(columns = {
    'brewery': 'count',
    'avg_rating': 'rating'
}, inplace = True)
```

```python
brew_more_5_sorted
```

```python
brew_more_5_sorted.reset_index(inplace = True)
brew_more_5_sorted
```

```python
top_5 = breweries_more_than_5[breweries_more_than_5.brewery.isin(brew_more_5_sorted['brewery'].head())].reset_index().drop('index', axis = 1)
top_5
```

```python
bottom_5 = breweries_more_than_5[breweries_more_than_5.brewery.isin(brew_more_5_sorted['brewery'].tail())].reset_index().drop('index', axis = 1)
bottom_5
```

```python
top_5_stats = top_5.groupby(['brewery', 'style']).aggregate({'brewery': 'count', 'avg_rating': 'mean'})
top_5_stats.rename(columns = {
    'brewery': 'count',
    'avg_rating': 'rating'
}, inplace = True)
top_5_stats.sort_values(by = ['brewery', 'count'], ascending = (1, 0), inplace = True)
top_5_stats.reset_index(inplace = True)
top_5_stats
```

```python
bottom_5_stats = bottom_5.groupby(['brewery', 'style']).aggregate({'brewery': 'count', 'avg_rating': 'mean'})
bottom_5_stats.rename(columns = {
    'brewery': 'count',
    'avg_rating': 'rating'
}, inplace = True)
bottom_5_stats.sort_values(by = ['brewery', 'count'], ascending = (1, 0), inplace = True)
bottom_5_stats.reset_index(inplace = True)
bottom_5_stats
```

```python
fig, axs = plt.subplots(2, 3, figsize = (24, 10))

for ax, (name, sub_df) in zip(axs.ravel(), top_5_stats.groupby('brewery')):
    labels = sub_df['rating']
    ax.barh(sub_df['style'], sub_df['count'])
    ax.set_title(name)
    ax.set_xticks(range(0, 14, 1))

fig.delaxes(axs[1][2])
fig.tight_layout()
```

```python
fig, axs = plt.subplots(2, 3, figsize = (24, 10))

for ax, (name, sub_df) in zip(axs.ravel(), bottom_5_stats.groupby('brewery')):
    ax.barh(sub_df['style'], sub_df['count'])
    ax.set_title(name)
    ax.set_xticks(range(0, 14, 1))

fig.delaxes(axs[1][2])
fig.tight_layout()
```

**Question 2: Based on the text analysis, what are the two or three most frequently associated words for the top 10 (by rating) styles of beer?**

```python
style_more_than_20 = df.groupby('style').filter(lambda x : x['style'].size >= 20)
style_more_than_20
```

```python
top_10_styles = style_more_than_20.groupby('style').aggregate({'avg_rating': 'mean'}).sort_values(by = 'avg_rating', ascending = False).head(10)
top_10_styles.reset_index(inplace = True)
top_10_styles
```

```python
top_10 = style_more_than_20[style_more_than_20['style'].isin(top_10_styles['style'])].reset_index().drop('index', axis = 1)
top_10
```

```python
top_10_agg = top_10.groupby('style').aggregate({
    'text_astringency': 'sum',
    'text_body': 'sum',
    'text_alcohol': 'sum',
    'text_bitter': 'sum',
    'text_sweet': 'sum',
    'text_sour': 'sum',
    'text_salty': 'sum',
    'text_fruits': 'sum',
    'text_hoppy': 'sum',
    'text_spices': 'sum',
    'text_malty': 'sum'
})
top_10_agg.reset_index(inplace = True)
top_10_agg.rename(columns = ({
    'style': 'beer_style',
    'text_astringency': 'astringency',
    'text_body': 'body',
    'text_alcohol': 'alcohol',
    'text_bitter': 'bitter',
    'text_sweet': 'sweet',
    'text_sour': 'sour',
    'text_salty': 'salty',
    'text_fruits': 'fruits',
    'text_hoppy': 'hoppy',
    'text_spices': 'spices',
    'text_malty': 'malty'}), inplace = True)
top_10_agg
```

```python
top_10_melt = top_10_agg.melt(id_vars = 'beer_style', 
                          value_vars = ['astringency', 'body', 'alcohol', 'bitter', 'sweet', 
                                        'sour', 'salty', 'fruits', 'hoppy', 'spices', 'malty'],
                          var_name = 'attrib', value_name = 'count') 
top_10_sort = top_10_melt.sort_values(by = ['beer_style', 'count'], ascending = (1, 0))
top_10_attrib = top_10_sort.groupby('beer_style').head(3).reset_index().drop('index', axis = 1)
top_10_attrib
```

```python
fig, axs = plt.subplots(2, 5, figsize = (24, 10))

for ax, (name, sub_df) in zip(axs.ravel(), top_10_attrib.groupby('beer_style')):
    ax.pie(sub_df['count'], labels = sub_df['attrib'], autopct='%.2f%%')
    ax.set_title(name)

fig.tight_layout()
```
